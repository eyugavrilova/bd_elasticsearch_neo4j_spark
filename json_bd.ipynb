```python
import json
import random
from datetime import datetime, timedelta
```

```python
# Количество генерируемых баннеров
pacient_numb = 30

# Количество генерируемых кликов
procedure_numb = 30
```

Баннер:
{index, doc_type, id, body: {дата_заказа, заказчик, рекламная_компания, требования_к_размещению*,  ссылка_на_изображение, ранг, [клиент*]}}
Клик (по баннеру):
{index, doc_type, id, body: {id_баннера, дата,  IP_клиента, URL_сайта, [UTM_метка]}}
Примечание. Квадратные скобки [] обозначает тег (может быть несколько значений)
поля, отмеченные *, разделить на слова, убрать пунктуацию с помощью токенизатора standart (русский), перевести все токены в нижний регистр, убрать токены, находящиеся в списке стоп-слов, выполнить стемминг оставшихся токенов с помощью фильтра snowball

Пациент:
{index, doc_type, id, body: {id_пациента, персональные_данные*, путёвка, дата_прибытия, продолжительность_прибывания, [диагноз*], [id_процедуры]}}
Процедура:
{index, doc_type, id, body: {сведения_о_процедуре*,  стоимость}}
Примечание. Квадратные скобки [] обозначает тег (может быть несколько значений)
поля, отмеченные *, разделить на слова, убрать пунктуацию с помощью токенизатора standart (русский), перевести все токены в нижний регистр, убрать токены, находящиеся в списке стоп-слов, выполнить стемминг оставшихся токенов с помощью фильтра snowball.


```python
#путёвка
tours = ['Ночной дятел',
         'Шерстяной волчара',
         'Азимут',
         'Золотой лось',
         'Территория странствий',
         'Турмарафон',
         'ММБ',
         'Радиалка',
         'Подкустовный выползень',
         'Звёздный ворс']
#продолжительность пребывания
duration = ['7 дней', '14 дней', '21 день', '28 дней']

#персональные данные
city = [
    'Балакирево',
    'Воинова Гора',
    'Арсаки',
    'Донховка',
    'Абрамцево',
    'Жилёво',
]

gender = ['Женщина', 'Мужчина']
age = [
    'Младше 18 лет',
    '18 - 25 лет',
    '26 - 30 лет',
    '31 - 40 лет',
    '41 - 50 лет',
    'Старше 50 лет']

#диагноз
diagnosis = [
    'Хроническая усталость от жизни',
    'Острая нехватка новых впечатлений',
    'Гиподинамия',
    'Обострение офисного рабства',
    'Весна в душе',
    'Наличие острого инструмента сапожника в мягких тканях задней и латеральной поверхностей таза'
]

#сведения о процедуре
procedure_info = [
    'Грязевые ванны. Продолжительность - 24 часа, после перерыва - ещё 12 часов. ',
    'Хоровое пение и медитации. ',
    'Прогревание у костра. ',
    'Микстура "каркаде богов", 0.3 - 0.5л. ',
    'Бороплюс животворящий, местно. ',
    'Тигровый бальзам. ',
    'Воздушные ванны. ',
    'Прослушивание расслабляющего пения птиц. ',
    'Прогревание организма изнутри. ',
    'Сенсорная депривация путём погружения в спальный мешок. ',
    'Ингаляции дымом костра. ',
    'Ароматерапия запахом осенней листвы. ',
    'Солнечные ванны. ',
    'Болотные ванны. ',
    'Дождевые ванны. ',
    'Шоковая адреналинотерапия: убегание от стада кабанчиков. ',
    'Тейпирование всего организма. ',
    'Спортивная ходьба с подволакиванием обеих ног. ',
    'Тренировка слуха путём угадывания расположения станции на КП по характерному писку. ',
    'Тренировка зрения путём разглядывания звёзд',
    'Распаривание ног. ',
    'Болотно-грязевые компрессы. ',
    'Укрепляющие заплывы через ручьи. ',
    'Развитие мелкой моторики путём сбора ягод на дистанции. ',
    'Диетический стол: гречка с тушёнкой. ',
    'Выпрямление позвоночника: сон на земле с подложенной пенкой. ',
    'Восстановление режима дня: подъём в 6 утра. ',
    'Неспешная вечерная пробежка на 70 км. ',
    'Цветотерапия цветущей медуницей. ',
    'Лечебный сон в электричке в течение 2 часов. '
]

#стоимость
cost = [
    200,
    500,
    700,
    1000,
    1200,
    1500
]

#дозировка
dosage = [
    'Разово.',
    'Ежедневно.',
    'Еженедельно.',
    'Два раза в день.',
    'По самочувствию.',
    'До полного выздоровления'
]

```

```python
# Название файла с пациентами
pacient_fname = 'pacients.json'

# Название файла с процедурами
proc_fname = 'procedures.json'
```

```python
# Возвращает случайную дату в течение 10 лет со start
def rand_date(start=datetime(2016, 2, 7)):
    end = start + timedelta(days=random.randint(0, 1825))
    return end.strftime('%Y-%m-%d')
```

```python
# случайный выбор путёвки
def rand_tour():
    return random.choice(tours)

# случайный выбор продолжительности
def rand_duration():
    return random.choice(duration)

# случайный выбор стоимости
def rand_cost():
    return random.choice(cost)

#id пациента
def rand_id_pacient():
    return random.randrange(1, 100, 1)

# диагноз
def rand_diagnosis():
    return random.sample(diagnosis, random.randint(1, len(diagnosis) - 1))

# id процедуры
def rand_id_proc():
    return random.sample(proc_id, random.randint(1, 10))

#сведения о процедуре
def rand_proc_info(i):
    return procedure_info[i] + random.choice(dosage)
```

```python
proc_id = []
for i in range (1,31):
    proc_id.append(i)
```

```python
# рандомные персональные данные
def rand_personal_data():
    return '. '.join([random.choice(age), random.choice(gender), random.choice(city)])
```

```python
# запись в файл данных
def json_output(data, file_name):
    with open(file_name, 'w') as json_file:
        json.dump(data, json_file, indent=4, sort_keys=True, ensure_ascii=False)


pacients = []
procedures = []
```

```python
for i in range(1, pacient_numb + 1):
    pacient = {
        'index': 'pacients',
        'id': i,
        'body': {
            'pacient_id': rand_id_pacient(),
            'personal_data': rand_personal_data(),
            'tour': rand_tour(),
            'arriving_date': rand_date(),
            'duration': rand_duration(),
            'diagnosis': rand_diagnosis(),
            'procedure_id': rand_id_proc()
        }
    }

    pacients.append(pacient)

```

```python
for i in range(1, procedure_numb + 1):
    
    procedure = {
        'index': 'procedures',
        'id': i,
        'body': {
            'procedure_info': rand_proc_info(i-1),
            'cost': rand_cost()
        }
    }

    procedures.append(procedure)
```

```python
for i in pacients:
     print(json.dumps(i, indent=4, sort_keys=True))

for j in procedures:
     print(json.dumps(j, indent=4, sort_keys=True))

```

```python
json_output(pacients, pacient_fname)
json_output(procedures, proc_fname)
```
